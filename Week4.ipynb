# **데이터 분석을 위한 판다스: M, N, O**

## M. 데이터 병합하기

```python
import pandas as pd
```

```python
# 실습을 위한 데이터 프레임 만들기 1
adf = pd.DataFrame({"x1" : ["A","B","C"], 
                    "x2": [1,2,3]})
adf
```

```python
# 데이터 프레임 만들기 2
bdf = pd.DataFrame({"x1" : ["A","B","D"], 
                    "x3": ["T","F","T"]})
bdf
```

```python
# 데이터 프레임 만들기 3
cdf = pd.DataFrame({"x1" : ["B","C","D"], 
                    "x2": [2,3,4]})
cdf
```

### 70. 왼쪽 열을 축으로 병합하기

```python
# x1을 키로 해서 병합, 왼쪽(adf)를 기준으로
# 왼쪽의 adf에는 D가 없으므로 해당 값은 NaN으로 변환합니다.
pd.merge(adf, bdf, how='left', on='x1')
```

### 71. 오른쪽 열을 축으로 병합하기

```python
# x1을 키로 해서 병합, 오른쪽(bdf)을 기준으로
# 오른쪽의 bdf에는 C가 없으므로 해당 값은 NaN으로 변환합니다.
pd.merge(adf, bdf, how='right', on='x1')
```

### 72. 공통의 값만 병합하기

```python
pd.merge(adf, bdf, how='inner', on='x1')
```

### 73. 모든 값을 병합하기

```python
pd.merge(adf, bdf, how='outer', on='x1')
```

### 74. 특정한 열을 비교하여 공통 값이 존재하는 경우만 가져오기

```python
# adf와 bdf의 특정한 열을 비교하여 공통 값이 존재하는 경우만 가져옵니다.
# adf.x1열과 bdf.x1열은 A,B가 같다. 따라서 adf의 해당 값만 출력합니다.
adf[adf.x1.isin(bdf.x1)]
```

### 75. 공통 값이 존재하는 경우 해당 값을 제외하고 병합하기

```python
adf[~adf.x1.isin(bdf.x1)]  # 해당 값만 빼고 출력합니다.
```

### 76. 공통의 값이 있는 것만 병합하기

```python
pd.merge(adf, cdf)
```

### 77. 모두 병합하기

```python
pd.merge(adf, cdf, how='outer')
```

### 78. 어디서 병합되었는지 표시하기

```python
pd.merge(adf, cdf, how='outer', indicator=True)
```

### 79. 원하는 병합만 남기기

```python
pd.merge(adf, cdf, how='outer', indicator=True).query('_merge == "left_only"')
```

### 80. merge 컬럼 없애기

```python
pd.merge(adf, cdf, how='outer', indicator=True).query('_merge == "left_only"').drop(columns=['_merge'])
```

## N.  데이터 가공하기

```python
df = pd.DataFrame(
        {"a" : [4 ,5, 6, 7],   # 열 이름을 지정해 주고 시리즈 형태로 데이터를 저장
        "b" : [8, 9, 10, 11],
        "c" : [12, 13, 14, 15]},
        index = [1, 2, 3, 4])  # 인덱스는 1,2,3으로 정해 줍니다.
df
```

### 81. 행 전체를 한 칸 아래로 이동하기

```python
df.shift(1)
```

### 82. 행 전체를 한 칸 위로 이동하기

```python
df.shift(-1)
```

### 83. 첫 행부터 누적해서 더하기

```python
df.cumsum()
```

### 84. 새 행과 이전 행을 비교하면서 최댓값 출력하기

```python
df.cummax()
```

### 85. 새 행과 이전 행을 비교하면서 최솟값 출력하기

```python
df.cummin()
```

### 85. 첫 행부터 누적해서 곱하기

```python
df.cumprod()
```

## O. 그룹별로 집계하기

```python
# 실습을 위해 데이터를 불러옵니다.
!git clone https://github.com/taehojo/data.git 

# 15장의 주택 가격 예측 데이터를 불러 옵니다.
df = pd.read_csv("./data/house_train.csv")
df
```

### 87. 그룹 지정 및 그룹별 데이터 수 표시

```python
df.groupby(by="YrSold").size()   # 팔린 연도를 중심으로 그룹을 만든 후 그룹별 수 표시
```

### 88. 그룹 지정 후 원하는 칼럼 표시하기

```python
# 팔린 연도를 중심으로 그룹을 만든 후 각 그룹별로 주차장의 넓이를 표시합니다.
df.groupby(by="YrSold")['LotArea'].mean()
```

### 89. 밀집도 기준으로 순위 부여하기

```python
df['SalePrice'].rank(method='dense')   # 각 집 값은 밀집도를 기준으로 몇 번째인가
```

### 90. 최젓값을 기준으로 순위 부여하기

```python
df['SalePrice'].rank(method='min')    # 각 집 값이 최젓값을 기준으로 몇 번째인가
```

### 91. 순위를 비율로 표시하기

```python
# 집 값의 순위를 비율로 표시합니다(0=가장 싼 집, 1=가장 비싼 집).
df['SalePrice'].rank(pct=True)
```

### 92. 동일 순위에 대한 처리 방법 정하기

```python
# 순위가 같을 때 순서가 빠른 사람을 상위로 처리합니다.
df['SalePrice'].rank(method='first')
```
